```python
# üì¶ Import necessary libraries
import rasterio                    # To read and write GeoTIFF raster files
from rasterio.plot import show    # Optional: to visually display the raster
import numpy as np                # To work with raster data as arrays
from scipy.ndimage import generic_filter  # To apply custom filters for filling missing data
import matplotlib.pyplot as plt   # For visualization (optional but helpful)
```

```python
file_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'
with rasterio.open(file_path) as src:
    raster_data = src.read(1).astype(float)  # Make sure it's float so we can store NaNs
    profile = src.profile

    nodata_value = src.nodata
    print("NoData value:", nodata_value)

    # Replace NoData values with NaN
    if nodata_value is not None:
        raster_data[raster_data == nodata_value] = np.nan
```

```python
# üí° Count total and NaN cells
total_cells = raster_data.size
nan_cells = np.count_nonzero(np.isnan(raster_data))
percent_nan = (nan_cells / total_cells) * 100

print(f"Missing (NaN) data: {percent_nan:.2f}% of total cells")
```

```python
# üõ†Ô∏è Define a custom function to replace NaNs with the mean of valid neighbors
def fill_nan_with_mean(window):
    valid_values = window[~np.isnan(window)]  # Get non-NaN values in the 3x3 window
    if len(valid_values) > 0:
        return np.mean(valid_values)         # Return average of neighbors
    else:
        return np.nan                        # If all are NaN, return NaN
```

```python
# üß™ Apply the filter over the raster (3x3 neighborhood around each cell)
filled_raster = generic_filter(raster_data, fill_nan_with_mean, size=3, mode='constant', cval=np.nan)

# üîÑ Merge: where original was NaN, replace with filled value; else keep original
raster_filled = np.where(np.isnan(raster_data), filled_raster, raster_data)
```

```python
# üìä Check results
total_cells = raster_filled.size
nan_cells = np.count_nonzero(np.isnan(raster_filled))
percent_nan = (nan_cells / total_cells) * 100
print(f"After filling, missing data is now: {percent_nan:.2f}%")
```

```python
# üì¶ Import libraries
import rasterio
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import generic_filter

# üì• Load the raster
file_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'
with rasterio.open(file_path) as src:
    raster_data = src.read(1)
    profile = src.profile

# üîç Step 1: Detect potential NoData placeholder values
unique_values = np.unique(raster_data)
print("üëÄ Unique values in raw raster:", unique_values[:15])
print("üìé NoData from metadata:", profile['nodata'])

# üõ† Step 2: Replace known invalid values with NaN
possible_invalids = [-9999, 0, -32768, 3.40282e+38, 1.70141e+38]
for val in possible_invalids:
    if val in unique_values:
        raster_data = np.where(raster_data == val, np.nan, raster_data)

# üßÆ Step 3: Initial missing data report
total_cells = raster_data.size
nan_cells = np.count_nonzero(np.isnan(raster_data))
print(f"üìä Initial missing (NaN) data: {nan_cells / total_cells * 100:.2f}%")

# üîÅ Step 4: Define nan-filling function (3x3 mean)
def fill_nan_internal(values):
    center = values[len(values) // 2]
    if np.isnan(center):
        valid = values[~np.isnan(values)]
        return np.mean(valid) if valid.size > 0 else np.nan
    else:
        return center

# üîÑ Step 5: Iteratively fill only **internal** NaNs
raster_filled = raster_data.copy()
max_iterations = 20
rows, cols = raster_filled.shape

for i in range(max_iterations):
    before = np.count_nonzero(np.isnan(raster_filled))
    
    # Apply filter to all
    filtered = generic_filter(raster_filled, fill_nan_internal, size=3, mode='mirror')
    
    # Only update inner cells (ignore borders)
    updated = raster_filled.copy()
    updated[1:-1, 1:-1] = np.where(
        np.isnan(raster_filled[1:-1, 1:-1]),
        filtered[1:-1, 1:-1],
        raster_filled[1:-1, 1:-1]
    )

    raster_filled = updated
    after = np.count_nonzero(np.isnan(raster_filled))
    
    print(f"üîÅ Iteration {i+1}: {before - after} cells filled, {after} remaining")

    if after == before:
        print("‚úÖ No more internal fills possible. Stopping early.")
        break

# üìâ Final report
final_nan_cells = np.count_nonzero(np.isnan(raster_filled))
print(f"üßπ Final missing data after filling: {final_nan_cells / total_cells * 100:.4f}%")

# üìä Visual comparison
plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.title("Original (with NaN)")
plt.imshow(raster_data, cmap='terrain')
plt.colorbar(label="Elevation")

plt.subplot(1, 2, 2)
plt.title("After Internal NaN Filling")
plt.imshow(raster_filled, cmap='terrain')
plt.colorbar(label="Elevation")

plt.tight_layout()
plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# üì• Load the raster
file_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'
with rasterio.open(file_path) as src:
    raster_data = src.read(1)
    profile = src.profile

# ‚úÖ Remove only known invalid placeholders ‚Äî leave 0 untouched for now
invalid_vals = [-9999, -32768, 3.40282e+38, 1.70141e+38]
for val in invalid_vals:
    raster_data = np.where(raster_data == val, np.nan, raster_data)

# üß† Create a mask of valid (non-zero, non-NaN) data
valid_mask = (raster_data != 0) & (~np.isnan(raster_data))

# üó∫Ô∏è Find bounding box of valid data
rows = np.any(valid_mask, axis=1)
cols = np.any(valid_mask, axis=0)
row_start, row_end = np.where(rows)[0][[0, -1]]
col_start, col_end = np.where(cols)[0][[0, -1]]

# üßº Replace only the outer zero-padding with NaN
cleaned = raster_data.copy()
cleaned[:row_start, :] = np.where(cleaned[:row_start, :] == 0, np.nan, cleaned[:row_start, :])
cleaned[row_end+1:, :] = np.where(cleaned[row_end+1:, :] == 0, np.nan, cleaned[row_end+1:, :])
cleaned[:, :col_start] = np.where(cleaned[:, :col_start] == 0, np.nan, cleaned[:, :col_start])
cleaned[:, col_end+1:] = np.where(cleaned[:, col_end+1:] == 0, np.nan, cleaned[:, col_end+1:])

# üìä Check how much was cleaned
total = cleaned.size
nan_count = np.count_nonzero(np.isnan(cleaned))
print(f"‚úÖ Cleaned outer zeros. Total NaN: {nan_count} cells ({nan_count/total*100:.2f}%)")
print(f"üìê Valid data area: Rows {row_start}‚Äì{row_end}, Cols {col_start}‚Äì{col_end}")

# üñºÔ∏è Visualize cleaned DEM
plt.figure(figsize=(8, 6))
plt.title("DEM (Zero-padding removed)")
plt.imshow(cleaned, cmap='terrain')
plt.colorbar(label="Elevation")
plt.tight_layout()
plt.show()
```

```python
import rasterio
import numpy as np
import matplotlib.pyplot as plt

# üì• Load the raster
file_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/slope.tif'
with rasterio.open(file_path) as src:
    raster_data = src.read(1)
    profile = src.profile

# üîß Step 1: Replace known invalids with NaN (but leave zero alone for now)
invalid_vals = [-9999, -32768, 3.40282e+38, 1.70141e+38]
for val in invalid_vals:
    raster_data = np.where(raster_data == val, np.nan, raster_data)

# üß† Step 2: Build mask of actual valid elevation values
# Anything non-zero AND non-NaN is valid
valid_mask = (~np.isnan(raster_data)) & (raster_data > 0)

# Find bounding box of valid area
rows = np.any(valid_mask, axis=1)
cols = np.any(valid_mask, axis=0)

try:
    row_start, row_end = np.where(rows)[0][[0, -1]]
    col_start, col_end = np.where(cols)[0][[0, -1]]
except IndexError:
    raise ValueError("‚ö†Ô∏è No valid data found in raster.")

# üí• Step 3: Replace everything outside this box with NaN ‚Äî regardless of value (including 0s)
cleaned = raster_data.copy()
cleaned[:row_start, :] = np.nan
cleaned[row_end+1:, :] = np.nan
cleaned[:, :col_start] = np.nan
cleaned[:, col_end+1:] = np.nan

# ‚úÖ Summary
total = cleaned.size
nan_count = np.count_nonzero(np.isnan(cleaned))
print(f"‚úÖ Padding removed. Total NaN: {nan_count} cells ({nan_count/total*100:.2f}%)")
print(f"üìê Valid data bounding box: Rows {row_start}‚Äì{row_end}, Cols {col_start}‚Äì{col_end}")

# üñºÔ∏è Visualize cleaned raster
plt.figure(figsize=(8, 6))
plt.title("DEM (Strict Padding Removed)")
plt.imshow(cleaned, cmap='terrain')
plt.colorbar(label="Elevation")
plt.tight_layout()
plt.show()
```

```python
# üì¶ Imports
import rasterio                           # For reading/writing geospatial rasters
import numpy as np                        # Core array operations
import matplotlib.pyplot as plt           # Visualization
from scipy.ndimage import label           # For connected region detection
from rasterio.enums import Resampling     # For possible rescaling

# üì• Load the DEM file
file_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'

with rasterio.open(file_path) as src:
    dem = src.read(1)                    # Read the first band of the raster
    profile = src.profile                # Store metadata (CRS, resolution, etc.)

# üîß Replace known invalid placeholders with np.nan
# Do NOT remove zero ‚Äî it‚Äôs valid (flat ground)
invalid_values = [-9999, -32768, 3.40282e+38, 1.70141e+38]
for val in invalid_values:
    dem = np.where(dem == val, np.nan, dem)

# üß† Build a mask of valid data (elevation > 0 and not NaN)
valid_mask = (~np.isnan(dem)) & (dem > 0)

# üß© Use connected-component labeling to find largest valid region (ignore scattered zones)
labeled_array, num_features = label(valid_mask)

# If no valid region found, stop
if num_features == 0:
    raise ValueError("‚ùå No valid terrain found in DEM.")

# Get the largest connected component only
sizes = [(labeled_array == i).sum() for i in range(1, num_features + 1)]
main_label = np.argmax(sizes) + 1
final_mask = (labeled_array == main_label)

# üßº Clean DEM: keep only largest terrain region, rest is set to NaN
dem_cleaned = np.where(final_mask, dem, np.nan)

# üîº Compute slope and aspect from the DEM using finite difference
# Simple 3x3 window gradient method (not GIS-precise but fast for ML)
xres = profile['transform'][0]
yres = -profile['transform'][4]

# Pad edges for safe diff computation
dem_pad = np.pad(dem_cleaned, 1, mode='edge')

# Compute gradients
dzdx = (dem_pad[1:-1, 2:] - dem_pad[1:-1, :-2]) / (2 * xres)
dzdy = (dem_pad[2:, 1:-1] - dem_pad[:-2, 1:-1]) / (2 * yres)

# Slope in degrees
slope = np.degrees(np.arctan(np.sqrt(dzdx**2 + dzdy**2)))

# Aspect in degrees
aspect = np.degrees(np.arctan2(dzdy, -dzdx))
aspect = np.where(aspect < 0, 90.0 - aspect, 450.0 - aspect)
aspect = np.where(np.isnan(dem_cleaned), np.nan, aspect)

# üß™ Sanity check: confirm shape match
assert dem_cleaned.shape == slope.shape == aspect.shape, "‚ùå Mismatched shapes in output rasters."

# üìä Report missing data %
total_pixels = dem_cleaned.size
nan_pixels = np.count_nonzero(np.isnan(dem_cleaned))
print(f"üìä DEM valid area: {(100 - nan_pixels / total_pixels * 100):.2f}%")

# üìà Plot DEM
plt.figure(figsize=(8, 6))
plt.title("Cleaned DEM")
plt.imshow(dem_cleaned, cmap='terrain')
plt.colorbar(label="Elevation (m)")
plt.tight_layout()
plt.show()

# üìà Plot Slope
plt.figure(figsize=(8, 6))
plt.title("Slope (degrees)")
plt.imshow(slope, cmap='viridis')
plt.colorbar(label="Slope (¬∞)")
plt.tight_layout()
plt.show()

# üìà Plot Aspect
plt.figure(figsize=(8, 6))
plt.title("Aspect (degrees)")
plt.imshow(aspect, cmap='twilight')
plt.colorbar(label="Aspect (¬∞)")
plt.tight_layout()
plt.show()
```

```python
import geopandas as gpd
import rasterio
from rasterio.features import rasterize
import numpy as np
from shapely.geometry import box

# üìÇ Paths
gpkg_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/firepack.gpkg'
layer_name = 'firepoints'
dem_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'
output_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/processed/firepack_buffered.npy'

# üì• Step 1: Load fire polygons
gdf = gpd.read_file(gpkg_path, layer=layer_name)

# üì• Step 2: Load DEM metadata
with rasterio.open(dem_path) as src:
    dem_shape = (src.height, src.width)
    dem_transform = src.transform
    dem_crs = src.crs  # Usually EPSG:4326

# ‚úÖ Step 3: Reproject fire polygons to UTM Zone 43N (EPSG:32643)
gdf_projected = gdf.to_crs("EPSG:32643")

# üîÅ Step 4: Buffer geometries outward (in meters)
buffer_distance =800  # You can try 200/300 for stronger effect
gdf_projected['geometry'] = gdf_projected.geometry.buffer(buffer_distance)

# üîÑ Step 5: Reproject back to DEM's CRS
gdf_buffered = gdf_projected.to_crs(dem_crs)

# üß± Step 6: Rasterize the buffered polygons
buffered_raster = rasterize(
    [(geom, 1) for geom in gdf_buffered.geometry],
    out_shape=dem_shape,
    transform=dem_transform,
    fill=0,
    dtype='uint8'
)

# üíæ Step 7: Save the final raster label
np.save(output_path, buffered_raster)

# üìä Step 8: Print summary
print("‚úÖ Buffered fire label saved to:", output_path)
print("üìê Raster shape:", buffered_raster.shape)
print("üî• Fire pixels after buffer:", np.count_nonzero(buffered_raster))
```

```python
import geopandas as gpd
import rasterio
import matplotlib.pyplot as plt
from rasterio.plot import show
from shapely.geometry import box

# üìÇ Paths
gpkg_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/firepack.gpkg'
layer_name = 'firepoints'
dem_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'

# üì• Load DEM raster
with rasterio.open(dem_path) as src:
    dem_data = src.read(1)
    dem_bounds = src.bounds
    dem_crs = src.crs
    dem_transform = src.transform

# üî≤ Create DEM boundary polygon for plotting
dem_box = gpd.GeoDataFrame({'geometry': [box(*dem_bounds)]}, crs=dem_crs)

# üì• Load original fire polygons
gdf = gpd.read_file(gpkg_path, layer=layer_name)

# üîÅ Reproject fire polygons to UTM for buffering
gdf_projected = gdf.to_crs("EPSG:32643")

# üìè Apply buffer
buffer_distance = 800  # Try 150 / 300 / 400 and compare
gdf_projected['geometry'] = gdf_projected.geometry.buffer(buffer_distance)

# üîÅ Reproject buffered fire polygons back to DEM CRS (usually EPSG:4326)
gdf_buffered = gdf_projected.to_crs(dem_crs)

# üñºÔ∏è Plot DEM + fire overlay
fig, ax = plt.subplots(figsize=(10, 10))
show(dem_data, transform=dem_transform, ax=ax, cmap='terrain')
gdf_buffered.boundary.plot(ax=ax, edgecolor='red', linewidth=1, label='Buffered Fire Zones')
plt.title(f"Fire Patches after {buffer_distance}m Buffer on DEM")
plt.legend()
plt.axis('off')
plt.show()
```
