```python
# üì¶ Imports
import rasterio                           # For reading/writing geospatial rasters
import numpy as np                        # Core array operations
import matplotlib.pyplot as plt           # Visualization
from scipy.ndimage import label           # For connected region detection
from rasterio.enums import Resampling     # For possible rescaling
import os
```

```python

# üì• Load the DEM file
file_path = 'C:/Users/Siddharth Tomar/forest-fire-simulation/data/raw/dem_raw.tif'

with rasterio.open(file_path) as src:
    dem = src.read(1)                    # Read the first band of the raster
    profile = src.profile                # Store metadata (CRS, resolution, etc.)
```

```python

# üîß Replace known invalid placeholders with np.nan
# Do NOT remove zero ‚Äî it‚Äôs valid (flat ground)
invalid_values = [-9999, -32768, 3.40282e+38, 1.70141e+38]
for val in invalid_values:
    dem = np.where(dem == val, np.nan, dem)
```

```python

# üß† Build a mask of valid data (elevation > 0 and not NaN)
valid_mask = (~np.isnan(dem)) & (dem > 0)

# üß© Use connected-component labeling to find largest valid region (ignore scattered zones)
labeled_array, num_features = label(valid_mask)
```

```python

# If no valid region found, stop
if num_features == 0:
    raise ValueError("‚ùå No valid terrain found in DEM.")

# Get the largest connected component only
sizes = [(labeled_array == i).sum() for i in range(1, num_features + 1)]
main_label = np.argmax(sizes) + 1
final_mask = (labeled_array == main_label)

# üßº Clean DEM: keep only largest terrain region, rest is set to NaN
dem_cleaned = np.where(final_mask, dem, np.nan)
```

```python

# üîº Compute slope and aspect from the DEM using finite difference
# Simple 3x3 window gradient method (not GIS-precise but fast for ML)
xres = profile['transform'][0]
yres = -profile['transform'][4]

# Pad edges for safe diff computation
dem_pad = np.pad(dem_cleaned, 1, mode='edge')

# Compute gradients
dzdx = (dem_pad[1:-1, 2:] - dem_pad[1:-1, :-2]) / (2 * xres)
dzdy = (dem_pad[2:, 1:-1] - dem_pad[:-2, 1:-1]) / (2 * yres)
```

```python

# Slope in degrees
slope = np.degrees(np.arctan(np.sqrt(dzdx**2 + dzdy**2)))

# Aspect in degrees
aspect = np.degrees(np.arctan2(dzdy, -dzdx))
aspect = np.where(aspect < 0, 90.0 - aspect, 450.0 - aspect)
aspect = np.where(np.isnan(dem_cleaned), np.nan, aspect)
```

```python

# üß™ Sanity check: confirm shape match
assert dem_cleaned.shape == slope.shape == aspect.shape, "‚ùå Mismatched shapes in output rasters."

# üìä Report missing data %
total_pixels = dem_cleaned.size
nan_pixels = np.count_nonzero(np.isnan(dem_cleaned))
print(f"üìä DEM valid area: {(100 - nan_pixels / total_pixels * 100):.2f}%")
```

```python

# üìà Plot DEM
plt.figure(figsize=(8, 6))
plt.title("Cleaned DEM")
plt.imshow(dem_cleaned, cmap='terrain')
plt.colorbar(label="Elevation (m)")
plt.tight_layout()
plt.show()
```

```python

# üìà Plot Slope
plt.figure(figsize=(8, 6))
plt.title("Slope (degrees)")
plt.imshow(slope, cmap='viridis')
plt.colorbar(label="Slope (¬∞)")
plt.tight_layout()
plt.show()
```

```python

# üìà Plot Aspect
plt.figure(figsize=(8, 6))
plt.title("Aspect (degrees)")
plt.imshow(aspect, cmap='twilight')
plt.colorbar(label="Aspect (¬∞)")
plt.tight_layout()
plt.show()
```
